```python
# Librerías
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
%matplotlib inline

# Inizializando la figura
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')

# Creando los datos x,y,z
z=np.linspace(-5,5,1000)
r=z**2+1
th=np.linspace(-8*np.pi,8*np.pi,1000)
x=r*np.cos(th)
y=r*np.sin(th)

# La gráfica
ax.plot(x,y,z)
```
